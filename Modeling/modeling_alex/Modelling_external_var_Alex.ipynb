### Prophet with weather and recursive lags.

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Best parameters
best_params = {'changepoint_prior_scale': 0.1, 'seasonality_prior_scale': 0.01, 'holidays_prior_scale': 0.01, 'seasonality_mode': 'additive'}

# Prepare the data for Prophet
df_prophet_train = df_train.reset_index()
df_prophet_train.columns = ['ds', 'y', 'temp_max_c']

df_prophet_test = df_test.reset_index()
df_prophet_test.columns = ['ds', 'y', 'temp_max_c']

# Create lag features on the training data
for lag in [1, 2, 7, 30, 60, 90, 365]:
    df_prophet_train[f'y_lag{lag}'] = df_prophet_train['y'].shift(lag)

# Drop rows with NaN values in the lag columns
df_prophet_train = df_prophet_train.dropna()

# Train the final model with the best parameters
model = Prophet(**best_params)
model.add_regressor('temp_max_c')
for lag in [1, 2, 7, 30, 60, 90, 365]:
    model.add_regressor(f'y_lag{lag}')
model.fit(df_prophet_train)

# Initialize future DataFrame for predictions
future = df_prophet_test.copy()

# Initialize lag features for future dataframe using the last known values from the training set
for lag in [1, 2, 7, 30, 60, 90, 365]:
    future[f'y_lag{lag}'] = df_prophet_train[f'y_lag{lag}'].iloc[-1]

# Make iterative predictions to fill lag features in the test set
predictions = []
for i in range(len(df_prophet_test)):
    forecast_data = model.predict(future.iloc[:i + 1])
    yhat = forecast_data['yhat'].iloc[-1]
    predictions.append(yhat)
    future.loc[i, 'y'] = yhat
    for lag in [1, 2, 7, 30, 60, 90, 365]:
        if i - lag >= 0:
            future.loc[i, f'y_lag{lag}'] = future.loc[i - lag, 'y']
        else:
            future.loc[i, f'y_lag{lag}'] = df_prophet_train.iloc[-lag]['y']

# Convert predictions to a DataFrame
predictions_df = pd.DataFrame({'ds': df_prophet_test['ds'], 'yhat': predictions})

# Combine train and test data for plotting
combined_df = pd.concat([df_prophet_train[['ds', 'y']], df_prophet_test[['ds', 'y']]])
combined_df = combined_df.set_index('ds')

# Plot the results using Prophet's built-in plotting functions
forecast_data = model.predict(future)

fig = model.plot(forecast_data)
plt.scatter(pd.to_datetime(df_prophet_test['ds']), pd.to_numeric(df_prophet_test['y']), color='r', marker='.')
plt.show()

# Plot components
fig_components = model.plot_components(forecast_data)
plt.show()

# Prepare the data for metrics calculation
y_train = df_train['rides'].values
y_test = df_test['rides'].values
y_pred = predictions_df['yhat'].values

# Calculate metrics
mae_value = mean_absolute_error(y_test, y_pred)
mse_value = mean_squared_error(y_test, y_pred)
mape_value = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
smape_value = 100/len(y_test) * np.sum(2 * np.abs(y_pred - y_test) / (np.abs(y_test) + np.abs(y_pred)))
mase_value = mae_value / np.mean(np.abs(np.diff(y_train)))
r2_value = r2_score(y_test, y_pred)

# Print metrics
print("MAE:", mae_value)
print("MSE:", mse_value)
print("MAPE:", mape_value)
print("SMAPE:", smape_value)
print("MASE:", mase_value)
print("R^2:", r2_value)
```

### LSTM with temp

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt

# Prepare the data
X_train = df_train[['temp_max_c']].values
y_train = df_train['rides'].values
X_test = df_test[['temp_max_c']].values
y_test = df_test['rides'].values

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Reshape the data for LSTM (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Function to create the LSTM model
def create_model(units=50, activation='relu', optimizer='adam'):
    model = Sequential()
    model.add(LSTM(units, activation=activation, input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(Dense(1))
    model.compile(optimizer=optimizer, loss='mse')
    return model

# Define the parameter grid with fewer parameters
param_grid = {
    'units': [50, 100],
    'activation': ['relu'],
    'optimizer': ['adam'],
    'epochs': [50],
    'batch_size': [16, 32]
}

# Function to perform grid search
def grid_search(param_grid, X_train, y_train, X_test, y_test):
    best_params = None
    best_score = float('inf')

    for units in param_grid['units']:
        for activation in param_grid['activation']:
            for optimizer in param_grid['optimizer']:
                for epochs in param_grid['epochs']:
                    for batch_size in param_grid['batch_size']:
                        model = create_model(units=units, activation=activation, optimizer=optimizer)
                        model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)
                        y_pred = model.predict(X_test).flatten()
                        mse = mean_squared_error(y_test, y_pred)
                        if mse < best_score:
                            best_score = mse
                            best_params = {
                                'units': units,
                                'activation': activation,
                                'optimizer': optimizer,
                                'epochs': epochs,
                                'batch_size': batch_size
                            }
    return best_params

# Perform grid search
best_params = grid_search(param_grid, X_train, y_train, X_test, y_test)
print("Best parameters:", best_params)

# Train the final model with the best parameters
model_lstm_best = create_model(units=best_params['units'], activation=best_params['activation'], optimizer=best_params['optimizer'])
model_lstm_best.fit(X_train, y_train, epochs=best_params['epochs'], batch_size=best_params['batch_size'], verbose=0)

# Make predictions
y_pred = model_lstm_best.predict(X_test).flatten()

# Calculate metrics
mae_value = mean_absolute_error(y_test, y_pred)
mse_value = mean_squared_error(y_test, y_pred)
mape_value = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
smape_value = np.mean(2 * np.abs(y_test - y_pred) / (np.abs(y_test) + np.abs(y_pred))) * 100

# Calculate MASE
naive_forecast = np.roll(y_train, shift=1)
naive_forecast[0] = y_train[0]
mase_value = np.mean(np.abs(y_test - y_pred)) / np.mean(np.abs(y_train - naive_forecast))

r2_value = r2_score(y_test, y_pred)

# Print metrics
print("MAE:", mae_value)
print("MSE:", mse_value)
print("MAPE:", mape_value)
print("SMAPE:", smape_value)
print("MASE:", mase_value)
print("R^2:", r2_value)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df_train.index, df_train['rides'], label='Train')
plt.plot(df_test.index, df_test['rides'], label='Test')
plt.plot(df_test.index, y_pred, label='Predicted')
plt.legend()
plt.show()
```
